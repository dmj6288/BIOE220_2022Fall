```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import scipy
```

```python
def pend(y, t, k1, gamma, kv):
    dydt = [-k1*y[0], gamma*y[0] - kv*y[1]]
    return dydt
```

```python
y0 = [20, 40.0]
k1 = 0.5
gamma = 2
kv = 0.5
t = np.linspace(0, 25, 1001)
sol = odeint(pend, y0, t, args=(k1, gamma, kv))
```

```python
plt.plot(t, sol.T[0])
plt.plot(t, sol.T[1])
plt.xlabel('Time')
plt.ylabel('Number')
plt.title('Equation solutions for Infected cell and virus particles')
plt.legend(['Infected cells', 'Viral particles'])
```
